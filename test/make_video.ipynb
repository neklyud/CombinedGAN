```python
import numpy as np
import cv2
import glob
import os
```

```python
cvs = sorted(glob.glob('./_1_*'), key = lambda x: int(x[5:-4]))
svs = sorted(glob.glob('./_2_*'), key = lambda x: int(x[5:]))
evs = sorted(glob.glob('./_3_*'), key = lambda x: int(x[5:-4]))
```

```python
svs
```

```python
img_array = []
for i in range(390):
    cvs_i = cv2.imread(cvs[i])
    svs_i = cv2.imread(svs[i])
    evs_i = cv2.imread(evs[i])
    conc = cv2.hconcat([evs_i, svs_i, cvs_i])
    height, width, layers = conc.shape
    size = (width,height)
    img_array.append(conc)
```

```python
out = cv2.VideoWriter('project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
